```python
#Gerekli Kütüphanerlerin Eklenmesi
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import numpy as np
```

### Girdi ve Çıktı Değişkenlerinin Tanımlanması 

```python
#Girdi Ve Çıktı Değerleri Kalite Sınıf Aralıkları ile Tanımlanır
shs = ctrl.Antecedent(np.arange(4,7,0.1),'shs')
tbs = ctrl.Antecedent(np.arange(3.5,8,0.1),'tbs')
prt = ctrl.Antecedent(np.arange(2,5,0.1),'prt')
qua = ctrl.Consequent(np.arange(0,1,0.05),'qua')
```

### Üyelik Fonksiyonlarının Tanımlanması 

```python
#Somatik Hücre Sayısının Üyelik Fonksiyonları
shs['yüksek'] = fuzzy.trapmf(shs.universe,[4,4,5.6,5.7])
shs['orta'] = fuzzy.trimf(shs.universe,[5.6,5.7,6])
shs['düşük'] = fuzzy.trapmf(shs.universe,[5.7,6,7,7])
#Toplam Bakteri Sayısının Üyelik Fonksiyonları
tbs['yüksek'] = fuzzy.trapmf(tbs.universe,[3.5,3.5,4.6,5])
tbs['orta'] =  fuzzy.trimf(  tbs.universe,[4.6,5,5.4])
tbs['düşük'] = fuzzy.trapmf( tbs.universe,[5,5.4,8,8])
#Protein Değerinin Üyelik Fonksiyonları
prt['düşük'] = fuzzy.trapmf(prt.universe,[2,2,2.5,2.7])
prt['orta'] =  fuzzy.trimf( prt.universe,[2.5,2.7,4.5])
prt['yüksek'] = fuzzy.trapmf( prt.universe,[2.7,4.5,5,5])
#Çiğ Süt Kalite Değerinin  Üyelik Fonksiyonları
qua['düşük'] =  fuzzy.trapmf(qua.universe,[0,0,0.25,0.4])
qua['orta'] =   fuzzy.trapmf(qua.universe,[0.25,0.4,0.6,0.75])
qua['yüksek'] =  fuzzy.trapmf(qua.universe,[0.6,0.75,1,1])
```

### Grafik

```python
#Grafikler
shs.view()
tbs.view()
prt.view()
qua.view()
```

### Kural Tabanı

```python
#Kurallar
csr_list = []
rule_1 = ctrl.Rule(tbs['yüksek']&shs['yüksek']&prt['yüksek'],[qua['düşük']])
csr_list.append(rule_1)
rule_2 = ctrl.Rule(tbs['yüksek']&shs['orta']&prt['yüksek'],  [qua['düşük']])
csr_list.append(rule_2)
rule_3 = ctrl.Rule(tbs['yüksek']&shs['düşük']&prt['yüksek'], [qua['düşük']])
csr_list.append(rule_3)
rule_4 = ctrl.Rule(tbs['yüksek']&shs['yüksek']&prt['orta'],  [qua['düşük']])
csr_list.append(rule_4) 
rule_5 = ctrl.Rule(tbs['yüksek']&shs['orta']&prt['orta'],    [qua['düşük']])
csr_list.append(rule_5)
rule_6 = ctrl.Rule(tbs['yüksek']&shs['düşük']&prt['orta'],   [qua['düşük']])
csr_list.append(rule_6)
rule_7 = ctrl.Rule(tbs['yüksek']&shs['yüksek']&prt['düşük'], [qua['düşük']])
csr_list.append(rule_7)
rule_8 = ctrl.Rule(tbs['yüksek']&shs['orta']&prt['düşük'],   [qua['düşük']])
csr_list.append(rule_8)
rule_9 = ctrl.Rule(tbs['yüksek']&shs['düşük']&prt['düşük'],  [qua['düşük']])
csr_list.append(rule_9)
rule_10 =  ctrl.Rule(tbs['orta']&shs['yüksek']&prt['yüksek'], [qua['düşük']])
csr_list.append(rule_10)
rule_11 =  ctrl.Rule(tbs['orta']&shs['orta']&prt['yüksek'],   [qua['orta']])
csr_list.append(rule_11)
rule_12 =  ctrl.Rule(tbs['orta']&shs['düşük']&prt['yüksek'],  [qua['orta']])
csr_list.append(rule_12)
rule_13 =  ctrl.Rule(tbs['orta']&shs['yüksek']&prt['orta'],   [qua['orta']])
csr_list.append(rule_13)
rule_14 = ctrl.Rule(tbs['orta']&shs['orta']&prt['orta'],     [qua['orta']])
csr_list.append(rule_14)
rule_15 = ctrl.Rule(tbs['orta']&shs['düşük']&prt['orta'],    [qua['orta']])
csr_list.append(rule_15)
rule_16 = ctrl.Rule(tbs['orta']&shs['yüksek']&prt['düşük'],  [qua['orta']])
csr_list.append(rule_16)
rule_17 = ctrl.Rule(tbs['orta']&shs['orta']&prt['düşük'],    [qua['orta']])
csr_list.append(rule_17)
rule_18 = ctrl.Rule(tbs['orta']&shs['düşük']&prt['yüksek'],  [qua['orta']])
csr_list.append(rule_18)
rule_19 = ctrl.Rule(tbs['düşük']&shs['yüksek']&prt['yüksek'],[qua['orta']])
csr_list.append(rule_19)
rule_20 = ctrl.Rule(tbs['düşük']&shs['orta']&prt['yüksek'],  [qua['yüksek']])
csr_list.append(rule_20)
rule_21 = ctrl.Rule(tbs['düşük']&shs['düşük']&prt['yüksek'], [qua['yüksek']])
csr_list.append(rule_21)
rule_22 = ctrl.Rule(tbs['düşük']&shs['yüksek']&prt['orta'],  [qua['orta']])
csr_list.append(rule_22)
rule_23 = ctrl.Rule(tbs['düşük']&shs['orta']&prt['orta'],    [qua['yüksek']])
csr_list.append(rule_23)
rule_24 = ctrl.Rule(tbs['düşük']&shs['düşük']&prt['orta'],   [qua['yüksek']])
csr_list.append(rule_24)
rule_25 = ctrl.Rule(tbs['düşük']&shs['yüksek']&prt['düşük'], [qua['orta']])
csr_list.append(rule_25)
rule_26 = ctrl.Rule(tbs['düşük']&shs['orta']&prt['düşük'],   [qua['yüksek']])
csr_list.append(rule_26)
rule_27 = ctrl.Rule(tbs['düşük']&shs['düşük']&prt['düşük'],  [qua['yüksek']])
csr_list.append(rule_27)
```

### Kural Tabanının Kontrol Sistemine Gönderilmesi 

```python
#Kurallar formatı değiştirilerek tanımlanır
qua_ctrl = ctrl.ControlSystem(csr_list)
#Kural Listesi Hesapnlanılarak Tanımlanır
qual = ctrl.ControlSystemSimulation(qua_ctrl)
qual1 = ctrl.ControlSystemSimulation(qua_ctrl)
qual2 = ctrl.ControlSystemSimulation(qua_ctrl)
```

### Durulaştırma

```python
#Durulaştırma
def get_linguist(valmf,val,val_range):
    mf_keys = list(valmf.terms.keys())
    mf_value_list = []
    for i in range(len(mf_keys)):
        mf_value_list.append(skfuzzy.interp_membership(val_range,valmf[mf_keys[i]].mf,val))
    max_mf = max(mf_value_list)
    ling_key = ''
    for i in range(len(mf_keys)):
        if max_mf == mf_value_list[i]:
            ling_key = mf_keys[i]
            break
    return ling_key   
```

## Testler

### Test 1 - Yüksek

```python
import skfuzzy
#Hesaplanmış listelere test edilecek değerlerin atılıp denenmesi
def test(qual,tbs,shs,prt):
    qual.input['tbs']  = tbs
    qual.input['shs']  = shs
    qual.input['prt']  = prt
    qual.compute()
#Grafikleri çizer
def draw(qual):
    tbs.view(sim = qual)
    shs.view(sim = qual)
    prt.view(sim = qual)
#Girilen değerlerin kalite sınıf aralıklarına göre kalite değerini yazar
def write(tbs_value,shs_value,prt_value):
    print("Toplam Bakteri Sayısı :",get_linguist(tbs,tbs_value,tbs_range))
    print("Somatik Hücre Sayısı  :",get_linguist(shs,shs_value,shs_range))
    print("Protein               :",get_linguist(prt,prt_value,prt_range))

shs_range = np.arange(4,7,0.1)
tbs_range = np.arange(3.5,8,0.1)
prt_range = np.arange(2,5,0.1)
qua_range = np.arange(0,1,0.05)    
```

```python
test(qual,6.5,6.6,4.8)
write(6.5,6.6,4.8)
```

```python
draw(qual)
```

```python
import skfuzzy
print("Kalite Değeri :",qual.output['qua'])
print("Çıktısı :",get_linguist(qua,qual.output['qua'],np.arange(0,1,0.05)))
qua.view(sim = qual)
```
